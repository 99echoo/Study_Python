```python
'''
2839번
상근이는 요즘 설탕공장에서 설탕을 배달하고 있다. 상근이는 지금 사탕가게에 설탕을 정확하게 N킬로그램을 배달해야 한다. 설탕공장에서 만드는 설탕은 봉지에 담겨져 있다. 봉지는 3킬로그램 봉지와 5킬로그램 봉지가 있다.

상근이는 귀찮기 때문에, 최대한 적은 봉지를 들고 가려고 한다. 예를 들어, 18킬로그램 설탕을 배달해야 할 때, 3킬로그램 봉지 6개를 가져가도 되지만, 5킬로그램 3개와 3킬로그램 1개를 배달하면, 더 적은 개수의 봉지를 배달할 수 있다.

상근이가 설탕을 정확하게 N킬로그램 배달해야 할 때, 봉지 몇 개를 가져가면 되는지 그 수를 구하는 프로그램을 작성하시오.
'''

map = [0, 2, 4, 1, 3]

N = int(input())

i = map[N%5]

j = (N - 3*i) // 5

print(i+j if j >= 0 else -1)
```

```python
'''
우선 N을 3과 5의 합으로 나타낼 수 있다고 가정해봅시다(즉 답이 -1이 아니라고 가정). 그리고 'N을 5로 나눈 나머지'라는 말을 '5를 최대 횟수로 사용해 N을 만들어보려 했지만 결국 남은 수'로 조금 풀어서 생각해봅시다. 그렇다면 우리는 다음과 같이 생각해볼 수 있습니다.

 

N을 5로 나눈 나머지가 0인 경우: N을 만들 때 5를 덜 사용하지 않으면, N까지 도달하는 데 0 + 5 * 0 = 0 = 3 * 0만큼이 더 필요합니다.
N을 5로 나눈 나머지가 1인 경우: N을 만들 때 5를 한 번 덜 사용하면, N까지 도달하는 데 1 + 5 * 1 = 6 = 3 * 2만큼이 더 필요합니다.
N을 5로 나눈 나머지가 2인 경우: N을 만들 때 5를 두 번 덜 사용하면, N까지 도달하는 데 2 + 5 * 2 = 12 = 3 * 4만큼이 더 필요합니다.
N을 5로 나눈 나머지가 3인 경우: N을 만들 때 5를 덜 사용하지 않으면, N까지 도달하는 데 3 + 5 * 0 = 3 = 3 * 1만큼이 더 필요합니다.
N을 5로 나눈 나머지가 4인 경우: N을 만들 때 5를 한 번 덜 사용하면, N까지 도달하는 데 4 + 5 * 1 = 9 = 3 * 3만큼이 더 필요합니다.
(자세히 보시면 3 * k에서의 k가 차례대로 map에 담겨 있음을 알 수 있습니다)

 

즉 위 과정은 N까지 부족한 수치를 3의 k배수로 커버할 수 있을 때까지 5를 덜 사용해보면서, 적절한 k값을 찾는 것입니다.

 

봉투 수를 최소화하기 위해서는 3을 5번 이상 활용할 일이 없다는 점(3을 다섯 번 더한다 == 5를 세 번 더한다), 그리고 5를 최대한 많이 사용하고, 남은 부분을 3으로 커버해야 한다는 점을 상기해봅시다. 이를 고려할 때, 결국 map에는 N을 5로 나눈 나머지가 q일 때 필요한 3의 최소 개수 map[q]가 담겨 있는 것이고, N이 3 * map[q]보다 작아 음수가 나온다면 도출한 규칙상 3과 5의 조합으로 N을 만들어내는  것이 불가능하므로 -1을 출력하게 되는 것입니다.
'''
```

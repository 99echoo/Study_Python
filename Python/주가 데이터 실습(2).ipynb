```python
import yfinance as yf
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

stock = 'Hmotors'
#code = '005380.KS'
code = '084990.KQ'

start = '2017-01-04'
end = '2019-01-05'
#SKT = yf.download('017670.KS', '2020-11-01', '2020-11-13')
#print(SKT)

data1 = yf.download(code, start, end)
data1['Close_return'] = data1['Close'].pct_change()
print(data1.head())

data1.dropna(axis=0, inplace=True)
sns.displot( data1['Close_return'], bins=100, kde=True)

plt.xlabel( 'Daily return')
plt.ylabel( 'Number of days')
plt.title('Distribution of '+stock+ ' daily returns between '+start+' and '+end)

plt.show()
```

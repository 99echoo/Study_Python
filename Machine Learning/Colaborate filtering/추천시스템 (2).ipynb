```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

rawRatings = pd.read_csv('C:/Users/user/Desktop/2023-2/추천시스템/DongHyun_HW1/ratings.csv')
rawMovies = pd.read_csv('C:/Users/user/Desktop/2023-2/추천시스템/DongHyun_HW1/movies.csv')

rawData = pd.merge(rawRatings, rawMovies, on='movieId')
df = pd.DataFrame(rawData)


UserBased_RatingMatrix = df.pivot_table('rating', columns ='title',index= 'userId')
UserBased_RatingMatrix.fillna(0,inplace = True)



def performItemBasedCF(ratingMatrix,targetUserIndex,N,K):

    itemSimilarityMatrix = cosine_similarity(ratingMatrix.T)
    np.fill_diagonal(itemSimilarityMatrix,0)


    lstSimilarity = itemSimilarityMatrix[targetUserIndex] ## 각 개인별 코사인 유사도를 뽑아내는 코드
    lstSimilarUsers = np.argsort(lstSimilarity)[::-1][:K] ##이걸 내림차순으로 k개만큼 뽑아준다.
    ###유사한 영화만큼 뽑았음
    
    recommendations = []
    for movieIndex, movieTitle in enumerate(ratingMatrix.columns):
        # 이미 본 영화는 제외
        if ratingMatrix.iat[targetUserIndex, movieIndex] == 0:
            # 가중평균으로 예상 평점 계산
            valueSum = 0
            weightSum = 0
            for neighbor in lstSimilarUsers:
                valueSum += ratingMatrix.iat[targetUserIndex, neighbor] * lstSimilarity[neighbor]
                weightSum += lstSimilarity[neighbor]
            if weightSum > 0:
                recommendations.append((movieTitle, valueSum / weightSum))


  
    # 평점을 기준으로 추천 영화를 내림차순으로 정렬
    recommendations.sort(key=lambda x: x[1], reverse=True)
    top_n_recommendations = recommendations[:N]

    return top_n_recommendations

import random

targetUserIndex = random.randint(0,609) # 대상 사용자 인덱스 (0 ~ 609)
N = 10   # 추천할 영화 수
K = 5    # 유사한 사용자의 수

top_n_recommendations = performItemBasedCF(UserBased_RatingMatrix, targetUserIndex, N, K)

# 사용자에게 추천하는 영화 목록을 출력
print(f"사용자 {targetUserIndex + 1}을 위한 상위 {N}개 영화 추천:")
for movie_title, predicted_rating in top_n_recommendations:
    print(f"{movie_title} (예상 평점: {predicted_rating:.2f})")
```

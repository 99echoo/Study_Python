```python
import random 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Q = 65
profit_list = []

for i in range(1000000):
    D = random.randint(40,70)
    if D > Q :
         D = Q

    SalesTotal = 2.5 * D
    PurchaseTotal = 1.5 * Q
    SalvageTotal = 0.5 * (Q - D)
    Profit = SalesTotal  - PurchaseTotal + SalvageTotal
    profit_list.append(Profit)
    
profit_simuls = pd.DataFrame()
profit_simuls['Profit'] = profit_list


plt.hist(profit_simuls['Profit'],bins=20)
plt.title('Magazine Boy Problem')
plt.xlabel('Profit')
plt.ylabel('Count')
plt.show()
print(profit_simuls['Profit'].describe())
```

```python
import random 
import numpy as np
import pandas as pd
Profit_list = []
Order = []

for Q in range(40,71) :
    
    for i in range(10000):
        D = random.randint(40,70)
        if D > Q :
             D = Q

        SalesTotal = 2.5 * D
        PurchaseTotal = 1.5 * Q
        SalvageTotal = 0.5 * (Q - D)
        Profit = SalesTotal  - PurchaseTotal + SalvageTotal 
        Profit_list.append(Profit)
        Order.append(Q)
        
profit_simuls = pd.DataFrame({'Profit' : Profit_list, 'Order' : Order})
        
ProfitMean_Std = profit_simuls.groupby('Order').aggregate(['mean','std'])
print(ProfitMean_Std)        
```
